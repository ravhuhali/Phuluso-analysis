```python


threshold = 0.7

row_remove = []
for col in df_cor.columns:
    for row in df_cor.index:
        if col != row and abs(df_cor.loc[row, col]) > threshold:
            print(f"{row} and {col}  {df_cor.loc[row, col]:.3f}")
            
            if row not in row_remove:

                row_remove.append(col)

```
